```python
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt



im = cv2.imread('lena.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
print(im.shape)
plt.figure(figsize=(20,20))
plt.imshow(im)
plt.show()

im_R = im.copy()
im_R[:, :, (1, 2)] = 0
im_G = im.copy()
im_G[:, :, (0, 2)] = 0
im_B = im.copy()
im_B[:, :, (0, 1)] = 0


plt.imshow(im_R)
plt.show()
plt.imshow(im_G)
plt.show()
plt.imshow(im_B)

plt.show()
```
