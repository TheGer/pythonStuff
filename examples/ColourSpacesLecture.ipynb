```python
import cv2 #open cv
import matplotlib.pyplot as plt
import matplotlib.cm as cm

image = cv2.imread('lena.jpg') #this will be in BGR format

#you can convert it to hsv
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
plt.show()
# note that we cannot display and hsv image in rgb

#display original image. NOTE: matplot lib uses RGB (not BGR). so:
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(rgb)
plt.show()

#convert to grayscale
gs = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#to display, you need to advise that it's a grayscale image
plt.imshow(gs, cmap=cm.gray, vmin=0, vmax=255)
plt.show()
```
