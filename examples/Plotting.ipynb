```python


import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1,2)

a = np.array([5,6, 22, 87, 5, 78, 32, 25, 83, 51, 75, 83, 89, 91])
bins = [0, 25, 50, 75, 100]

ax[0].hist(a, bins)
ax[0].set_xticks([0, 25, 50, 75, 100])
ax[0].set_xlabel("Grades")
ax[0].set_ylabel("No of students")
ax[0].set_title("Exam Results")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2



gray_img = cv2.imread('lena.jpg', 0) #load rgb image

plt.imshow(gray_img, cmap='gray')
plt.show()
#plt.hist(gray_img, 256)

#plt.show()

hist = cv2.calcHist(gray_img, [0], None, [256], [0,256])
plt.plot(hist)
plt.show()

gray_norm = cv2.equalizeHist(gray_img)
plt.imshow(gray_norm, cmap='gray')
plt.show()

hist = cv2.calcHist(gray_norm, [0], None, [256], [0,256])
plt.plot(hist)
plt.show()

```

Histogram using calcHist on gray channels

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2



rgb_img = cv2.imread('../images/lena.jpg') #load rgb image
gray_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.show()

hist = cv2.calcHist([gray_img], [0], None, [256], [0,256])
plt.plot(hist)
plt.show()
```

Histogram using calcHist on r g b  channels

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2



bgr_img = cv2.imread('../images/lena.jpg') #load rgb image
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()

hist = cv2.calcHist([rgb_img], [0], None, [256], [0,256])
plt.plot(hist, color=[1,0,0])
hist = cv2.calcHist([rgb_img], [1], None, [256], [0,256])
plt.plot(hist, color=[0,1,0])
hist = cv2.calcHist([rgb_img], [2], None, [256], [0,256])
plt.plot(hist, color=[0,0,1])
plt.show()
```
