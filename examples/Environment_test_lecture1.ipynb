```python

im = cv2.imread('lena.jpg')
plt.imshow(im)  #image read as BGR being displayed incorrectly as RGB
plt.show()

#convert BGR to RGB before display
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.show()
```
