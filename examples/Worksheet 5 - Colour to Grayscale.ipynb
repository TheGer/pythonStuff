```python
#method 1
import cv2 #open cv
import matplotlib.pyplot as plt;
import matplotlib.cm as cm;

img = cv2.imread('lena.jpg') #load rgb image
gs1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gs1, cmap=cm.gray, vmin=0, vmax=255)
plt.show()

#method 2
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
gs2 = hsv[:,:,2]
plt.imshow(gs2, cmap=cm.gray, vmin=0, vmax=255)
plt.show()

#method 3
gs3 = 0.298936 * img[:,:,2] + 0.587043 * img[:,:,1]  + 0.114021* img[:,:,0]
plt.imshow(gs3, cmap=cm.gray, vmin=0, vmax=255)
plt.show()
```
