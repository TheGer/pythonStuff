```python
# scaling, translation, rotation, reflection (flipping)

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('lena.jpg')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
#Original Image:
plt.title("Original")
plt.show()

#resizing by a factor of 2
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
res = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)
plt.imshow(res)
plt.title("Resized by factor of 2")
plt.show()
#OR

#resizing by specifying new dimensions
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
res = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)
plt.imshow(res)
plt.title("Resized by specifying new dimensions")
plt.show()
```

```python
import cv2
import numpy as np

img = cv2.imread('lena.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('Translated',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('lena.jpg',0)
rows,cols = img.shape

#setting pivot point to centre and angle to 90 degrees
M = cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('Rotated',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

img = cv2.imread('lena.jpg',0)
flipVertical = cv2.flip(img, 0)
flipHorizontal = cv2.flip(img, 1)
flipBoth = cv2.flip(img, -1)
cv2.imshow('Flipped',flipBoth)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
