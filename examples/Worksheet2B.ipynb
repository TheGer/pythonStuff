Outside border /padding

```python
from PIL import Image
import matplotlib.pyplot as plt #to display image
import numpy as np

im = np.array(Image.open('lena.jpg'))
im[:20,:,:]= 0
im[-20:,:,:]= 0
im[:,:20,:]= 0
im[:,-20:,:]= 0

plt.imshow(im)
plt.show()
```

Image flipping using flip() method

```python
#4.	Image flipping
#Write a program which loads an RGB image and flips it both vertically and horizontally. Do this by using the colon notation. Display the new image.
import numpy as np
import matplotlib.pyplot as plt #to display image
from PIL import Image

im = np.array(Image.open('lena.jpg'))

flipped = np.flip(im, (0,1)) #image flipped on axis 0 (rows) and axis 1 (columns)
plt.imshow(flipped)
plt.show()
```
