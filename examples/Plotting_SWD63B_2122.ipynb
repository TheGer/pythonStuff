```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
x= np.array(['Jan', 'Feb', 'Mar', 'Apr'])
y= np.array([45, 78, 34, 23])

plt.bar(x, y)
plt.xlabel("Months")
plt.ylabel("Rainfall in mm")
plt.title("Rainfall")
plt.show()
```

```python
a = np.array([22, 87, 5, 78, 32, 25, 83, 38, 35, 40, 75, 92, 12, 63])

bins = [0, 25, 50, 75, 100]
plt.figure(figsize=(20,20))
plt.hist(a, bins)
plt.xlabel("Grades")
plt.ylabel("Number of students")
plt.title("Rainfall")
plt.show()
```

```python
im_bgr = cv2.imread('flower.png')
im_rgb = cv2.cvtColor(im_bgr, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(im_rgb, cmap='gray')
plt.show()
plt.xlim(0, 255)
plt.hist(im_rgb.ravel(), 256)
plt.show()
norm = np.zeros(shape=im_rgb.shape)
norm = cv2.normalize(im_rgb, norm, 0, 255, cv2.NORM_MINMAX)
plt.imshow(norm, cmap='gray')
plt.show()
plt.xlim(0, 255)
plt.hist(norm.ravel(), 256)
plt.show()
```

```python
im_gray = cv2.imread('lena.jpg', 0)
plt.figure(figsize=(10,10))
plt.imshow(im_gray, cmap='gray')
plt.show()
plt.xlim(0, 255)
plt.hist(im_gray.ravel(), 256)
plt.show()

im_equ = cv2.equalizeHist(im_gray)

plt.figure(figsize=(10,10))
plt.imshow(im_equ, cmap='gray')
plt.show()
plt.xlim(0, 255)
plt.hist(im_equ.ravel(), 256)
plt.show()
```

```python
im_gray = cv2.imread('lena.jpg', 0)
plt.figure(figsize=(10,10))
plt.imshow(im_gray, cmap='gray')
plt.show()

hist = cv2.calcHist(im_gray, [0], None, [256], [0,255])
plt.plot(hist)
plt.show()
```

```python
im_bgr = cv2.imread('goldfish.jpg')
im_rgb = cv2.cvtColor(im_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
plt.show()

hist = cv2.calcHist([im_rgb], [0], None, [256], [0,256])
plt.plot(hist,  color='red')
hist = cv2.calcHist([im_rgb], [1], None, [256], [0,256])
plt.plot(hist,  color='green')
hist = cv2.calcHist([im_rgb], [2], None, [256], [0,256])
plt.plot(hist,  color='blue')
plt.show()
```
