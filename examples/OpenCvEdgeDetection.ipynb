```python
# and canny edge detection in the easiest way possible,
# using OpenCV

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('lena.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gaussian = cv2.GaussianBlur(gray,(3,3),0)

#canny
img_canny = cv2.Canny(img,150,200)

#sobel
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely

#laplacian
img_laplacian = cv2.Laplacian(img_gaussian,cv2.CV_8U)

original_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(original_rgb)
#Original Image:
plt.title("original")
plt.show()

canny_rgb = cv2.cvtColor(img_canny, cv2.COLOR_BGR2RGB)
plt.imshow(canny_rgb)
#Canny
plt.title("Canny")
plt.show()

sobelx_rgb = cv2.cvtColor(img_sobelx, cv2.COLOR_BGR2RGB)
plt.imshow(sobelx_rgb)
#Sobel x
plt.title("Sobel X")
plt.show()

sobely_rgb = cv2.cvtColor(img_sobely, cv2.COLOR_BGR2RGB)
plt.imshow(sobely_rgb)
#Sobel y
plt.title("Sobel Y")
plt.show()

sobel_rgb = cv2.cvtColor(img_sobel, cv2.COLOR_BGR2RGB)
plt.imshow(sobel_rgb)
#Sobel
plt.title("Sobel")
plt.show()

laplacian_rgb = cv2.cvtColor(img_laplacian, cv2.COLOR_BGR2RGB)
plt.imshow(laplacian_rgb)
#Laplacian
plt.title("Laplacian")
plt.show()
```
