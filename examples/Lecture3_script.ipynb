Display lena.jpg as separate R G B channels


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt




#doing some changes here 


imBGR = cv2.imread('lena.jpg')
plt.imshow(imBGR)
plt.show()

imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)

plt.imshow(imRGB)
plt.show()

#Display RED channel
im_R = imRGB.copy()
im_R[:, :, (1,2)] = 0
plt.imshow(im_R)
plt.show()

#Obtain and display Cyan
im_C = 255 - im_R
plt.imshow(im_C)
plt.show()

#amend code using above for G and B channels
im_G = imRGB.copy()
im_G[:, :-200, (0,2)] = 0
plt.imshow(im_G)
plt.show()

im_B = imRGB[:, :, (0, 1)] = 0
plt.imshow(im_B)
plt.show()
```
