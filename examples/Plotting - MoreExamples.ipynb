Example 2
Following example plots a histogram of marks obtained by students in a class. Four bins, 0-25, 26-50, 51-75, and 76-100 are defined. The Histogram shows number of students falling in this range.

```python
from matplotlib import pyplot as plt
import numpy as np
fig,ax = plt.subplots(1,1)
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
bins = [0,25,50,75,100]
plt.hist(a, bins)
plt.xlabel("Marks")
plt.ylabel("No of students")
plt.title('Exam Results')
plt.show()
```

Example 3
Set x-axis tick marks according to bins:

```python
from matplotlib import pyplot as plt
import numpy as np
fig,ax = plt.subplots(1,1)
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
bins = [0,25,50,75,100]
ax.hist(a, bins)
ax.set_title("histogram of result")
ax.set_xticks([0,25,50,75,100])
ax.set_xlabel('marks')
ax.set_ylabel('no. of students')
plt.show()
```

```python
# Color histogram OpenCV

import cv2

gfish = cv2.imread('goldfish.jpg')

gfish = cv2.cvtColor(gfish, cv2.COLOR_BGR2RGB)
plt.imshow(gfish)
plt.show()

#Plot color histogram excluding black due to high frequency
norm_img = cv2.normalize(gfish,  None, alpha = 0, beta = 255, norm_type = cv2.NORM_MINMAX, dtype = cv2.CV_32F)
for i, col in enumerate(['r', 'g', 'b']):
    hist = cv2.calcHist([norm_img], [i], None, [256], [1, 256]) #removed black frequency
    plt.plot(hist, color=col)
    plt.xlim([0, 256])

plt.show()


```

```python
#3D projection

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
from matplotlib import pyplot as plt
import numpy as np
import cv2

gfish = cv2.imread('goldfish.jpg')

gfish = cv2.cvtColor(gfish, cv2.COLOR_BGR2RGB)
plt.imshow(gfish)
plt.show()

r, g, b = cv2.split(gfish)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")

pixel_colors = gfish.reshape((np.shape(gfish)[0]*np.shape(gfish)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()

axis.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()
```

Visualising in HSV

```python
hsv_gfish = cv2.cvtColor(gfish, cv2.COLOR_RGB2HSV)

h, s, v = cv2.split(hsv_gfish)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")

axis.scatter(h.flatten(), s.flatten(), v.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Hue")
axis.set_ylabel("Saturation")
axis.set_zlabel("Value")
plt.show()

```
