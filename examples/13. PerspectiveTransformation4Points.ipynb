```python
# of a perspective transformation by knowing 4 points and their result
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('painting.jpg')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rows,cols,ch = img.shape

pts1 = np.float32([[390,130],[960,230],[370,720],[990,680]])
pts2 = np.float32([[0,0],[400,0],[0,300],[400,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(400,300))


plt.imshow(img)
plt.title('Input')
plt.show()
plt.imshow(dst)
plt.title('Output')
plt.show()
```
