```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('lena.jpg')
color = ('b','g','r') #blue is 0, 1 is green and 2 is red
#i is the number and col is the letter
for i,col in enumerate(color):
    print(i,col)
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()


histr = cv2.calcHist([img],[0],None,[256],[0,256])
histr2 = cv2.calcHist([img],[1],None,[256],[0,256])
plt.plot(histr,color = 'b')
plt.show()
plt.plot(histr2,color = 'g')
plt.xlim([0,300])

#how many pixels do we have at 150 for red?



```

Worksheet 7 - Task 2 (Histogram Equalisation):

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# reading an image using imreadmethod
my_img = cv2.imread('lena.jpg', 0) # ,0 converts image to grayscale
equ = cv2.equalizeHist(my_img)
# stacking both the images side-by-side orientation
res = np.hstack((my_img, equ))
# showing image input vs output
plt.imshow(res, cmap='gray')
plt.show()

#one interesting way of changing it would be to show each image separately without the hstack



print('Image histogram and equalised image histogram using matplotlib .hist')

fig,axs = plt.subplots(1,2)
axs[0].hist(my_img.ravel(), 256,[0,256])
axs[1].hist(equ.ravel(),256,[0,256])
plt.show()

print('Image histogram and equalised image histogram using opencv calcHist')
fig,axs = plt.subplots(1,2)
histcv2original = cv2.calcHist([my_img],[0],None,[256],[0,256])
histcv2equ = cv2.calcHist([equ],[0],None,[256],[0,256])
axs[0].plot(histcv2original)
axs[1].plot(histcv2equ)
plt.show()

print('Image histogram and normalised image histogram using opencv calcHist')

#first parameter: image, second parameter: image mask, third parameter: minimum, fourth parameter is the maximum
normalised = cv2.normalize(my_img,  None, 0, 100, cv2.NORM_MINMAX)


histcv2norm = cv2.calcHist([normalised],[0],None,[256],[0,256])
fig,axs = plt.subplots(1,2)
axs[0].plot(histcv2original)
axs[1].plot(histcv2norm)
plt.show()
```

Worksheet 8 - Task 3 (Histogram Normalization):

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# reading an image using imreadmethod
my_img = cv2.imread('locontrast.jpg') # ,0 converts image to grayscale
my_img = cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)

#normalizing the pictures based on a range of values between 0 and 255
final_img = cv2.normalize(my_img,  None, 0, 255, cv2.NORM_MINMAX)

#stack the images next to each other
norm = np.hstack((my_img, final_img))
# showing image input vs output
plt.imshow(norm)
plt.show()

fig,axs = plt.subplots(1,2)
axs[0].hist(my_img.ravel(), 256,[0,256])
axs[1].hist(final_img.ravel(),256,[0,256])
plt.show()

#use cv2 to generate the same histograms below the histograms shown above


```

Extra:  2D Histogram

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)
hist, xbins, ybins = np.histogram2d(h.ravel(),s.ravel(),[180,256],[[0,180],[0,256]])

plt.imshow(hist,interpolation = 'nearest')
plt.show()


```
