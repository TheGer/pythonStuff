Task 1

Demo version using iteration

```python
import cv2
import imutils
import matplotlib.pyplot as plt
import numpy as np

# load the image from disk

image = cv2.imread('trees.png')
# loop over the rotation angles
for angle in np.arange(0, 360, 15):
	rotated = imutils.rotate(image, angle)
	cv2.imshow("Rotated (Problematic)", rotated)
	cv2.waitKey(0)
# loop over the rotation angles again, this time ensuring
# no part of the image is cut off
for angle in np.arange(0, 360, 15):
	rotated = imutils.rotate_bound(image, angle)
	cv2.imshow("Rotated (Correct)", rotated)
	cv2.waitKey(0)
```

Student Worksheet version

```python
import cv2
import imutils
import matplotlib.pyplot as plt
import numpy as np

# load the image from disk as grayscale
image = cv2.imread('trees.png', 0)

#height, width = image.shape# keep original width
dim = (400, 400)

# resize image
resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)

print('Resized Dimensions : ',resized.shape)

rotated1 = imutils.rotate(resized, 30)
cv2.imshow("Rotated (Problematic)", rotated1)

rotated2 = imutils.rotate_bound(resized, 30)
cv2.imshow("Rotated (Correct)", rotated2)
cv2.waitKey(0)
```

To execute after Task 1

```python

t_value = int(input("What is the strength for image2 overlay on image1 (from 0 to 100) "))

img1_strength = 1 - t_value / 100
img2_strength = t_value / 100

super1 = cv2.addWeighted(rotated1, img1_strength, resized, img2_strength, 0)

plt.imshow(super1, cmap='gray')
plt.show()

```

TASK 3 - Resize to same

It prints every yth element from the list / array

a = [1,2,3,4,5,6,7,8,9]
a[::3]
[1, 4, 7]
The additional syntax of a[x::y] means get every yth element starting at position x

a[2::3]
[3, 6, 9]

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

src1 = cv2.imread('lena.jpg',0)
src2 = cv2.imread('rocket.jpg',0)

print(src1.shape)
print(src1.shape[1::-1])

src2 = cv2.resize(src2, src1.shape[1::-1])

plt.imshow(src1, cmap='gray')
plt.show()

plt.imshow(src2, cmap='gray')
plt.show()
```

TASK 4 - Concatenate

```python
import cv2
import numpy as np
from PIL import Image

src1 = cv2.imread('lena.jpg')
src2 = cv2.imread('rocket.jpg')
src1 = cv2.cvtColor(src1, cv2.COLOR_BGR2RGB)
src2 = cv2.cvtColor(src2, cv2.COLOR_BGR2RGB)

print(src1.shape)
print(src1.shape[1::-1])

src2 = cv2.resize(src2, src1.shape[1::-1])

panorama = np.concatenate((src1, src2), axis = 1)

plt.imshow(panorama)
plt.show()

img = Image.fromarray(panorama)
img.save('panorama.jpg')
```
