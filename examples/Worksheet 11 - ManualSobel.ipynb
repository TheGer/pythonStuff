```python
# algorithm manually, without using OpenCV

import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('lena.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

sobel_x = np.array([[ -1, 0, 1],
                    [ -2, 0, 2],
                    [ -1, 0, 1]])
sobel_y = np.array([[ -1, -2, -1],
                    [ 0, 0, 0],
                    [ 1, 2, 1]])

filtered_image_y = cv2.filter2D(gray, -1, sobel_y)
plt.imshow(filtered_image_y, cmap='gray')
plt.title("Sobel Y")
plt.show()

filtered_image_x = cv2.filter2D(gray, -1, sobel_x)
plt.imshow(filtered_image_x, cmap='gray')
plt.title("Sobel X")
plt.show()

filtered_image = filtered_image_y + filtered_image_x
plt.imshow(filtered_image, cmap='gray')
plt.title("Sobel")
plt.show()
```
