Question 1 : Generate a random grayscale image of 10x10 pixels,
with the following pixel values:
50,140,190


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

possible_values = np.array ([50,140,190], dtype ='uint8')

randomimg = np.random.choice(possible_values, size=[10,10])
print(randomimg)

#Note that lowest and highest values are being interpreted as black and white aka image is being normalised
plt.imshow(randomimg, cmap='gray')
plt.show()

#To set the limits from 0 to 255 use the following:
plt.imshow(randomimg, cmap='gray', vmin=0, vmax=255)
plt.show()

hist = cv2.calcHist([randomimg], [0], None, [256], [0,256])
plt.plot(hist)
plt.show()

equ = cv2.equalizeHist(randomimg)
histcv2equ = cv2.calcHist([equ],[0],None,[256],[0,256])
plt.plot(histcv2equ)
plt.show()
```
